```python
import numpy as np
import pandas as pd
```

```python
prices = pd.read_csv(r'S:\Coursera\Введение в машинное обучение\Неделя 4\Dow Jones\close_prices.csv', header=0)
dj = pd.read_csv(r'S:\Coursera\Введение в машинное обучение\Неделя 4\Dow Jones\djia_index.csv', header=0)
```

```python
prices = prices.iloc[:, 1:]
prices
```

```python
dj = dj.iloc[:, 1]
dj
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=10)
```

```python
pca.fit(prices)
```

```python
arr = pca.explained_variance_ratio_
arr
```

```python
count = 0
procent = 0
for i in arr:
    procent += i
    count +=1
    if procent > 0.90:
        break
print(count)
```

```python
print(sum(arr[0:4]))
```

```python
first_comp = pd.DataFrame(pca.transform(prices)[:,0])
first_comp
```

```python
coef = np.corrcoef(first_comp.T, dj.T)
coef
```

```python
print(prices.columns[np.argmax(pca.components_[0])])
```
